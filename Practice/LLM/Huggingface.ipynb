```python
bert_summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

# Example text to summarize
text = """
In the past decade, artificial intelligence (AI) has become one of the most important fields of research. It has led to innovations in many sectors, including healthcare, finance, and autonomous vehicles. AI systems have the potential to revolutionize the way we live and work. However, challenges such as ethics, bias, and transparency need to be addressed to fully realize its potential.
"""

# Generate a summary
summary = bert_summarizer(text)
print(summary)
```

```python
summary = bert_summarizer(text, max_new_tokens = 20)
print(summary)
```
